```python
 import pandas as pd
```

```python
df['gross'].fillna(df['gross'].mean(), inplace=True)
df['budget'].fillna(df['budget'].mean(), inplace=True)
df['content_rating'].fillna(df['content_rating'].mode()[0],inplace=True)
df['aspect_ratio'].fillna(df['aspect_ratio'].mean(),inplace=True)
df['title_year'].fillna(df['title_year'].mean(),inplace=True)
df['duration'].fillna(df['duration'].mean(),inplace=True)
df['num_critic_for_reviews'].fillna(df['num_critic_for_reviews'].mean(),inplace=True)
df['director_facebook_likes'].fillna(df['director_facebook_likes'].mean(),inplace=True)
df['plot_keywords'].fillna(df['plot_keywords'].mode()[0],inplace=True)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df = df.dropna()
```

```python
lower_thershold= []
upper_thershold= []
for i in ["num_critic_for_reviews", "duration", "director_facebook_likes","actor_3_facebook_likes","actor_1_facebook_likes","gross","num_voted_users","cast_total_facebook_likes","facenumber_in_poster","budget","actor_2_facebook_likes","aspect_ratio","movie_facebook_likes"]:
    IQR= df[i].quantile(0.75) - df[i].quantile(0.25)
    lower_bound= df[i].quantile(0.25) - (1.5*IQR)
    upper_bound= df[i].quantile(0.75) + (1.5*IQR)

    print(i, ":", lower_bound, ",",  upper_bound)

    lower_thershold.append(lower_bound)
    upper_thershold.append(upper_bound)
```
